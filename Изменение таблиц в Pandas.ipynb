```python
df = pd.read_csv('./laptop_price.csv', encoding="windows-1251", index_col='laptop_ID')
df.sample(5)
```

### Задание 1.1
Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)

```python
Cpu_Company=[]
for i, row in df.iterrows():
    Cpu_Company.append(row['Cpu'].split()[0])
df['Cpu_Company'] = Cpu_Company
df.sample(10)
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
df.sample(10)
```

### Задание 1.2
Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)

```python
def func(lst):
    sum_GB = 0 
    for el in lst.split():
        if 'GB' in el:
            sum_GB += float(el[:-2])
        elif 'TB' in el:
            sum_GB += float(el[:-2]) * 1024
    return int(sum_GB)

df['Memory_Amount'] = df.Memory.apply(func)
df.sample(10)
```

### Задание 1.3
Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)

```python
def memory_type(lst):
    memory = ' ' 
    for el in lst.split():
        if el.isalpha():
            if memory:
                memory += ' '
            memory += el
    return memory

df['Memory_Type'] = df.Memory.apply(memory_type)
df.sample(10)
```

### Задание 1.4
Удалите признаки Memory и ScreenResolution

```python
df.drop(columns = ['Memory', 'ScreenResolution'], inplace = True)
df.sample(10)
```

## Задание 4
Составьте несколько сводных таблиц


### Задание 4.1
Найдите среднюю стоимость ноутбуков в зависимости от компании производителя

Отсортируйте от меньшей стоимости к большей


```python
df.groupby('Company').agg({'Price_euros': 'mean'}).sort_values('Price_euros')
```

### Задание 4.2
Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора


```python
df.groupby('Cpu_Company').agg({'Price_euros': ['min', 'mean', 'max']})
```

### Задание 4.3
Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ


```python
df.pivot_table(index = 'Cpu_Company', columns = 'Ram', values = 'Weight', aggfunc = 'count', fill_value = 0)
```

### Задание 4.4
Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти


```python
df.pivot_table(index = 'OpSys', columns = 'Memory_Amount', values = 'Price_euros', aggfunc = 'mean', fill_value = 0)
```

```python
df.groupby(['OpSys', 'Memory_Amount']).agg({'Price_euros': 'mean'}).unstack('Memory_Amount').fillna(0)
```
